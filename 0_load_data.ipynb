```python
import urllib.request
import os
import geopandas as gpd
import shutil
from py7zr import unpack_7zarchive
```

Download and load the french data grid 

```python
url = 'https://www.insee.fr/fr/statistiques/fichier/4176290/Filosofi2015_carreaux_200m_shp.zip'
urllib.request.urlretrieve(url, 'sdf.zip')
shutil.unpack_archive('sdf.zip')
shutil.register_unpack_format('7zip', ['.7z'], unpack_7zarchive)
shutil.unpack_archive('Filosofi2015_carreaux_200m_metropole_shp.7z', '')
```

```python
sdf = gpd.read_file('Filosofi2015_carreaux_200m_metropole.shp')
```

```python
ax = sdf.plot(figsize=(10,10))
ax.set_axis_off()
```

Download the Quartiers prioritaires de la politique de la ville

```python
url = 'https://www.data.gouv.fr/fr/datasets/r/6da6f4de-aed4-4ab7-8ffc-7cb7ba1e6722'
urllib.request.urlretrieve(url, 'qpv.zip')
shutil.unpack_archive('qpv.zip')
```

```python
qpv = gpd.read_file('QP_METROPOLE_LB93.shp')
```

```python
ax = qpv[qpv.COMMUNE_QP =="Nantes"].plot(figsize=(10,10))
ax.set_axis_off()
```
